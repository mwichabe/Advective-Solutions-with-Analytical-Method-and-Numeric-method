```python
import numpy as np
import matplotlib.pyplot as plt

parameters = {
    "case1": {
        "mu1": 0.1,
        "mu2": 1.0,
        "k_rw": 1.0,
        "k_ro": 1.0,
        "a": 1.2,
        "b": 3,
        "S1i": 0,
        "S2r": 0.2,
        "Sv_star": 0.5,
        "c": 0.3,
        "Svi": 0.1
    },
    "case2": {
        "mu1": 1.0,
        "mu2": 0.1,
        "k_rw": 0.15,
        "k_ro": 1.0,
        "a": 4.0,
        "b": 1.5,
        "S1i": 0.2,
        "S2r": 0.5,
        "Sv_star": 0.5,
        "c": 0.3,
        "Svi": 0.15
    },
    "case3": {
        "mu1": 0.4,
        "mu2": 0.1,
        "k_rw": 0.4,
        "k_ro": 0.8,
        "a": 8,
        "b": 3,
        "S1i": 0.2,
        "S2r": 0.15,
        "Sv_star": 0.5,
        "c": 0.3,
        "Svi": 0.05
    }
}

def k_relative(Sv, Svi, Sv_star, S2r, k_rw, k_ro, a, b):
    S_sat = (Sv - Svi) / (1 - S2r - Svi)
    krw = k_rw * (S_sat ** a)
    kro = k_ro * ((1 - S_sat) ** b)
    return krw, kro

def capillary_pressure(Sv, Svi, Sv_star, S2r, c, tol=1e-6):
    delta_Sv = Sv_star - Svi

    if abs(delta_Sv) < tol:
        return float('inf')
    else:
        term1 = (Sv_star / delta_Sv) ** (1 / c)
        if c != 0:
            term2 = (Sv / delta_Sv) ** ( 1 / c)
        else:
            term2 = 1
        term3 = (Sv_star / delta_Sv) ** (1 / c) - 1
        Pc = c * term1 * term2 - term3
        return Pc




def analytical_solution_advective(S1i, S2r, parameters, vD):
    mu1 = parameters['mu1']
    mu2 = parameters['mu2']
    k_rw = parameters['k_rw']
    k_ro = parameters['k_ro']
    a = parameters['a']
    b = parameters['b']

    Sv = vD * S1i / (1 + (mu1 / mu2) * (k_ro / k_rw) * vD)
    return Sv

def numerical_solution_advective(S1i, S2r, parameters, vD_values):
    mu1 = parameters['mu1']
    mu2 = parameters['mu2']
    k_rw = parameters['k_rw']
    k_ro = parameters['k_ro']
    a = parameters['a']
    b = parameters['b']
    delta_t = 0.01
    delta_x = 0.1
    max_time = 1.0

    S1_values = np.zeros(len(vD_values))
    S1_values[0] = S1i
    for i in range(1, len(vD_values)):
        krw, kro = k_relative(S1_values[i-1], S1i, 0, S2r, k_rw, k_ro, a, b)
        F1 = krw / (mu1 / mu2 * kro + krw)
        S1_values[i] = S1_values[i-1] - delta_t / delta_x * (1 - F1) * (S1_values[i-1] - S2r)
        S1_values[i] = max(S1_values[i], S1i)

    return S1_values

def analytical_solution_imbibition(S1i, Sv_star, S2r, parameters, omega):
    mu1 = parameters['mu1']
    mu2 = parameters['mu2']
    k_rw = parameters['k_rw']
    k_ro = parameters['k_ro']
    a = parameters['a']
    b = parameters['b']
    c = parameters['c']

    Pc = capillary_pressure(S1i, S1i, Sv_star, S2r, c)
    t_star = (omega ** 2) * (mu1 + mu2) / (mu1 * mu2 * Pc)

    S1_values = np    .zeros(len(omega))

    for i, w in enumerate(omega):
        if w == 0:
            S1_values[i] = S2r
        else:
            S1_values[i] = S2r + (S1i - S2r) * (1 - (Pc / w) ** (1 / c))

    return S1_values, t_star

def numerical_solution_imbibition(S1i, Sv_star, S2r, parameters, omega_values):
    mu1 = parameters['mu1']
    mu2 = parameters['mu2']
    k_rw = parameters['k_rw']
    k_ro = parameters['k_ro']
    a = parameters['a']
    b = parameters['b']
    c = parameters['c']

    delta_omega = 0.01
    delta_t = 0.1
    max_time = 1.0
    S1_values = np.zeros(len(omega_values))
    S1_values[0] = S1i
    for i in range(1, len(omega_values)):
        Pc = capillary_pressure(S1_values[i-1], S1i, Sv_star, S2r, c)
        krw, kro = k_relative(S1_values[i-1], S1i, Sv_star, S2r, k_rw, k_ro, a, b)
        F1 = krw / (mu1 / mu2 * kro + krw)
        S1_values[i] = S1_values[i-1] + delta_t / delta_omega * (1 - F1) * (S2r - S1_values[i-1])

    return S1_values

def plot_advective_results(Sv_analytical, Sv_numerical):
    print("Analytical Solution:")
    print(Sv_analytical)

    print("Numerical Solution:")
    print(Sv_numerical)
    plt.figure(figsize=(10, 6))
    plt.plot(Sv_analytical, label='Analytical Solution')
    plt.plot(Sv_numerical, label='Numerical Solution')
    plt.xlabel('Index')
    plt.ylabel('Sv')
    plt.title('Comparison of Analytical and Numerical Solutions for Advective Flow')
    plt.legend()
    plt.grid(True)
    plt.show()


def plot_imbibition_results(Sv_analytical, Sv_numerical):
    plt.figure(figsize=(10, 6))
    plt.plot(Sv_analytical, label='Analytical Solution')
    plt.plot(Sv_numerical, label='Numerical Solution')
    plt.xlabel('Index')
    plt.ylabel('Sv')
    plt.title('Comparison of Analytical and Numerical Solutions for Spontaneous Imbibition')
    plt.legend()
    plt.grid(True)
    plt.show()


def main():
    for case, params in parameters.items():
        vD_values = np.linspace(0, 1, 100)

        Sv_analytical = analytical_solution_advective(params['S1i'], params['S2r'], params, vD_values)

        Sv_numerical = numerical_solution_advective(params['S1i'], params['S2r'], params, vD_values)

        plot_advective_results(Sv_analytical, Sv_numerical)

        omega_values = np.linspace(0, 1, 100)

        Sv_analytical_imbibition, t_star = analytical_solution_imbibition(params['S1i'], params['Sv_star'], params['S2r'], params, omega_values)

        Sv_numerical_imbibition = numerical_solution_imbibition(params['S1i'], params['Sv_star'], params['S2r'], params, omega_values)

        plot_imbibition_results(Sv_analytical_imbibition, Sv_numerical_imbibition)
        Sv_values = np.linspace(params['S2r'], params['S1i'], 100)
        Pc_values = [capillary_pressure(Sv, params['S1i'], params['Sv_star'], params['S2r'], params['c']) for Sv in Sv_values]
        plt.figure(figsize=(8, 6))
        plt.plot(Sv_values, Pc_values)
        plt.xlabel('Sv')
        plt.ylabel('Capillary Pressure')
        plt.title('Capillary Pressure Curve - Case {}'.format(case))
        plt.grid(True)
        plt.show()
        krw_values, kro_values = k_relative(Sv_values, params['S1i'], params['Sv_star'], params['S2r'], params['k_rw'], params['k_ro'], params['a'], params['b'])
        F1_values = krw_values / (params['mu1'] / params['mu2'] * kro_values + krw_values)
        plt.figure(figsize=(8, 6))
        plt.plot(Sv_values, F1_values)
        plt.xlabel('Sv')
        plt.ylabel('Fractional Flow')
        plt.title('Fractional Flow Curve - Case {}'.format(case))
        plt.grid(True)
        plt.show()


if __name__ == "__main__":
    main()

```
